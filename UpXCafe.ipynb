```python
import numpy as np
import pandas as pd
import json
from ipywidgets import interact
from bokeh.plotting import figure, show, ColumnDataSource
from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure
output_notebook()
```

```python
coffee = pd.read_json('D:/upxcafe/Cafe.json')
#coffee = coffee.sample(n=1000)
```

```python
stdf = coffee['Monthly Data'].apply(json.loads).apply(pd.Series)
```

```python
coffee = coffee[['Store ID', 'Name', 'Brand', 'Store Number', 'Phone Number',
       'Ownership Type', 'Street Combined', 'Street 1', 'Street 2', 'Street 3',
       'City', 'State', 'Country', 'Coordinates', 'Latitude', 'Longitude',
       'Start Date', 'Monthly Data']]
```

```python
#coffee.head()
```

```python
coffee_subset = coffee[['Name','Latitude','Longitude','Start Date']]
coffee_subset = coffee_subset.reset_index(drop=True)
coffee_subset = coffee_subset.sort_values(by='Start Date')
```

```python
St_dt = coffee_subset['Start Date'].unique()
len(St_dt)
```

```python
x = np.array(coffee_subset['Latitude'].where(coffee_subset['Start Date'] == St_dt[0]))
y = np.array(coffee_subset['Longitude'].where(coffee_subset['Start Date'] == St_dt[0]))

source = ColumnDataSource(data=dict(x=x, y=y))

plot = figure( plot_width=300, plot_height=300)

r = plot.scatter('x', 'y', source=source, line_width=10, line_alpha=0.6)
```

```python
def update(n):
    r.data_source.data['x'] = coffee_subset['Latitude'].where(coffee_subset['Start Date'] <= St_dt[n])
    r.data_source.data['y'] = coffee_subset['Longitude'].where(coffee_subset['Start Date'] <= St_dt[n])
    push_notebook()
```

```python
show(plot, notebook_handle=True)
```

```python
interact(update, n=(0,len(St_dt)-1))
```
